```python
#Feature Engineering-1 Assignment
"""Q1: What are missing values in a dataset? Why is it essential to handle missing values? Name some
algorithms that are not affected by missing values."""
Ans: Missing values in a dataset are values that are not present or have been removed from the dataset. It is essential to 
handle missing values because they can lead to inaccurate results and conclusions when analyzing the data. Algorithms that are
not affected by missing values include Decision Trees, Random Forests, and K-Nearest Neighbors.

"""Q2: List down techniques used to handle missing data. Give an example of each with python code."""
Ans:
1. Imputation: Imputation is a technique used to replace missing values with substituted values. For example, using the pandas 
library in Python, we can use the fillna() function to replace missing values with a given value: 

import pandas as pd 

df = pd.DataFrame({'A':[1,2,np.nan], 'B':[5,np.nan,np.nan], 'C':[1,2,3]}) 

df.fillna(0) 

2. Dropping: Dropping is a technique used to remove rows or columns with missing values. For example, using the pandas library
in Python, we can use the dropna() function to drop rows or columns with missing values: 

import pandas as pd 

df = pd.DataFrame({'A':[1,2,np.nan], 'B':[5,np.nan,np.nan], 'C':[1,2,3]}) 

df.dropna() 

3. Predictive Modeling: Predictive modeling is a technique used to predict missing values using existing data. For example, 
using the scikit-learn library in Python, we can use the KNeighborsRegressor class to predict missing values: 

from sklearn.neighbors import KNeighborsRegressor 

X = df[['A', 'B']] 
y = df['C'] 
knn = KNeighborsRegressor() 
knn.fit(X, y) 
predictions = knn.predict(X) 

4. End of Distribution Imputation: End of distribution imputation is a technique used to replace missing values with values at the 
far end of the distribution. For example, using the pandas library in Python, we can use the fillna() function to replace 
missing values with the maximum value in the column: 

import pandas as pd 

df = pd.DataFrame({'A':[1,2,np.nan], 'B':[5,np.nan,np.nan], 'C':[1,2,3]}) 

df.fillna(df.max())

"""Q3: Explain the imbalanced data. What will happen if imbalanced data is not handled?"""
Ans: Imbalanced data is a type of data where the classes are not represented equally. For example, in a dataset with two classes,
one class may have 90% of the data points while the other class has only 10%. This can lead to problems when training machine 
learning models because the model may be biased towards the majority class.

If imbalanced data is not handled, the model may be unable to accurately classify the minority class due to its lack of 
representation in the training data. This can lead to poor performance on unseen data and inaccurate predictions. Additionally,
it can lead to overfitting on the majority class, which can lead to poor generalization.

Therefore, it is important to handle imbalanced data by either oversampling the minority class or undersampling the majority 
class. This will ensure that the model is able to accurately classify both classes and generalize better on unseen data.

"""Q4: What are Up-sampling and Down-sampling? Explain with an example when up-sampling and down-
sampling are required."""

Ans: Up-sampling is the process of increasing the sampling rate of a signal by inserting zeros between samples. This increases 
the resolution of the signal and can be used to reduce aliasing. An example of when up-sampling is required is when a signal 
needs to be converted from one sampling rate to another.

Down-sampling is the process of decreasing the sampling rate of a signal by removing samples. This reduces the resolution of 
the signal and can be used to reduce noise. An example of when down-sampling is required is when a signal needs to be converted
from one sampling rate to another.

In both cases, up-sampling and down-sampling are required when a signal needs to be converted from one sampling rate to another.

"""Q5: What is data Augmentation? Explain SMOTE."""
Ans:Data Augmentation is a technique used to increase the size of a dataset by artificially generating new data points. It is 
used to improve the performance of machine learning models by providing more data for training.

SMOTE (Synthetic Minority Over-sampling Technique) is a data augmentation technique used to address the problem of imbalanced
datasets. It works by generating synthetic data points that are similar to existing minority class data points. This helps to
balance the dataset and improve the performance of machine learning models.

SMOTE works by randomly selecting a minority class data point and then finding its k nearest neighbors. It then generates a new
data point by taking the difference between the selected data point and its neighbors, and adding it to the original data point.
This process is repeated until the desired number of synthetic data points is generated.

"""Q6: What are outliers in a dataset? Why is it essential to handle outliers?"""
Ans: Outliers are data points that are significantly different from the rest of the data in a dataset. They can be caused by 
errors in data collection or measurement, or they can be legitimate values that are simply far away from the majority of the 
data. It is essential to handle outliers because they can have a significant impact on the results of any analysis performed
on the dataset. Outliers can skew the results and lead to incorrect conclusions. Therefore, it is important to identify and 
remove outliers from a dataset before performing any analysis.

"""Q7: You are working on a project that requires analyzing customer data. However, you notice that some of the data is missing.
What are some techniques you can use to handle the missing data in your analysis?"""

Ans:
1. Imputation: This involves replacing missing values with substituted values such as the mean, median, or mode of the 
available data.
2. Deletion: This involves removing records or variables with missing data from the analysis.
3. Model-based imputation: This involves using predictive models to estimate the missing values.
4. K-nearest neighbor imputation: This involves using the values from similar cases (based on distance measures) to fill in 
missing data.
5. Multiple imputation: This involves creating multiple datasets with different imputed values and then combining the results.
6. Hot deck imputation: This involves using the values from similar cases to fill in missing data.

"""Q8: You are working with a large dataset and find that a small percentage of the data is missing. What are some strategies 
you can use to determine if the missing data is missing at random or if there is a pattern to the missing data?"""
Ans: 1. Check the distribution of the missing data across different variables. If the missing data is randomly distributed, then it is likely to be missing at random.

2. Look for any correlations between the missing data and other variables in the dataset. If there is a pattern to the missing 
data, then it is likely to be missing not at random.

3. Use a statistical test such as a chi-square test to determine if the missing data is randomly distributed or if there is a 
pattern to the missing data.

4. Use multiple imputation techniques to fill in the missing data and then compare the results to the original dataset. 
If the results are similar, then it is likely that the missing data is missing at random.

"""Q9: Suppose you are working on a medical diagnosis project and find that the majority of patients in the dataset do not have
the condition of interest, while a small percentage do. What are some strategies you can use to evaluate the performance of 
your machine learning model on this imbalanced dataset?"""

Ans: 1. Use metrics such as precision, recall, and F1 score to evaluate the performance of your model. These metrics take into 
account both false positives and false negatives, which are important when dealing with imbalanced datasets.

2. Use oversampling or undersampling techniques to balance the dataset. Oversampling involves duplicating examples from the 
minority class, while undersampling involves removing examples from the majority class.

3. Use a cost-sensitive learning algorithm, which assigns different weights to different classes in order to account for the 
imbalance in the dataset.

4. Use a combination of precision-recall curves and ROC curves to evaluate the performance of your model. Precision-recall 
curves are better suited for imbalanced datasets, while ROC curves are better suited for balanced datasets.

"""Q10: When attempting to estimate customer satisfaction for a project, you discover that the dataset is unbalanced, with the 
bulk of customers reporting being satisfied. What methods can you employ to balance the dataset and down-sample the majority 
class?"""

A: Methods to balance the dataset and down-sample the majority class include:
1. Random undersampling: randomly select a subset of the majority class to reduce its size.
2. Synthetic minority oversampling technique (SMOTE): generate new samples of the minority class by interpolating between 
existing samples.
3. Adaptive Synthetic Sampling (ADASYN): generate new samples of the minority class by interpolating between existing samples, 
but with an emphasis on generating more samples from regions where the minority class is more heavily represented.
4. Cluster-based undersampling: group the majority class into clusters and then randomly select a subset of the clusters to 
reduce its size.
5. Tomek links: remove samples from the majority class that are close to samples from the minority class.

"""Q11: You discover that the dataset is unbalanced with a low percentage of occurrences while working on a project that 
requires you to estimate the occurrence of a rare event. What methods can you employ to balance the dataset and up-sample the 
minority class?"""

Ans: 
1. Random Oversampling: This method randomly replicates the minority class to balance the dataset.
2. Synthetic Minority Oversampling Technique (SMOTE): This technique creates synthetic data points of the minority class to 
balance the dataset.
3. Adaptive Synthetic Sampling (ADASYN): This technique creates synthetic data points of the minority class based on the 
distribution of the majority class.
4. Cluster-Based Over-Sampling: This technique creates clusters of the minority class and then randomly replicates them to 
balance the dataset.
5. Edited Nearest Neighbor (ENN): This technique removes data points from the majority class that are close to the minority 
class to balance the dataset.

```
